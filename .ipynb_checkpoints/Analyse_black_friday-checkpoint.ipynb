```python
import pandas as pd
bf_data = pd.read_csv("BlackFriday.csv")
import matplotlib.pyplot as plt
import numpy as np
```

```python
bf_data.info()
```

```python
plt.plot()
plt.draw()
```

```python
Age_group = bf_data.groupby("Age")
Age_group.count()
```

```python
achat_client = dict()
for age, group in Age_group:
    achat_client[age] = sum(group["Purchase"])
```

```python
achat_client
```

```python
plt.plot(achat_client.keys(), achat_client.values(), marker='o')
plt.draw()
```

```python
plt.plot(achat_client.keys(), achat_client.values(), marker='o', linestyle="")
plt.draw()
```

```python
plt.plot(achat_client.keys(), achat_client.values(), marker='o', linestyle="")
plt.xlabel("Category d'age")
plt.ylabel("Sommes total des ventes ($)")
plt.title("Sommes des ventes par categorie d'age")
plt.draw()
```

```python
plt.rcParams["figure.figsize"]=[16,9]
plt.scatter(achat_client.keys(), achat_client.values())
plt.xlabel("Category d'age")
plt.ylabel("Sommes total des ventes ($)")
plt.title("Sommes des ventes par categorie d'age")
plt.draw()
```

```python
plt.rcParams["figure.figsize"]=[16,9]
plt.scatter(achat_client.keys(), achat_client.values(), c=list(achat_client.values()), cmap="seismic")
plt.xlabel("Category d'age")
plt.ylabel("Sommes total des ventes ($)")
plt.title("Sommes des ventes par categorie d'age")
plt.colorbar(orientation="vertical")
plt.draw()
```

```python
plt.rcParams["figure.figsize"]=[16,9]
plt.scatter(achat_client.keys(), achat_client.values(), c=list(achat_client.values()), cmap="seismic", marker="^")
plt.xlabel("Category d'age")
plt.ylabel("Sommes total des ventes ($)")
plt.title("Sommes des ventes par categorie d'age")
plt.colorbar(orientation="vertical")
plt.draw()
```

```python
plt.rcParams["figure.figsize"]=[16,9]
plt.scatter(achat_client.keys(), achat_client.values(), c=list(achat_client.values()), cmap="seismic", marker="^", s=100)
plt.xlabel("Category d'age")
plt.ylabel("Sommes total des ventes ($)")
plt.title("Sommes des ventes par categorie d'age")
plt.colorbar(orientation="vertical")
plt.savefig("graphic.png", bbox_inches="tight")
plt.draw()
```

```python
plt.rcParams["figure.figsize"]=[16,9]
plt.bar(achat_client.keys(), achat_client.values())
plt.xlabel("Category d'age")
plt.ylabel("Sommes total des ventes ($)")
plt.title("Sommes des ventes par categorie d'age")
plt.savefig("graphic.png", bbox_inches="tight")
plt.draw()
```

```python
plt.rcParams["figure.figsize"]=[16,9]
plt.bar(achat_client.keys(), achat_client.values(), color=["b", "g", "r", "c", "m", "y", "g"])
plt.xlabel("Category d'age")
plt.ylabel("Sommes total des ventes ($)")
plt.title("Sommes des ventes par categorie d'age")
plt.savefig("graphic.png", bbox_inches="tight")
plt.draw()
```

```python
achat_genre = bf_data["Gender"].value_counts()
achat_genre
```

```python
plt.bar(achat_genre.index, achat_genre, color=["b", "r"])
plt.xlabel("Sexe")
plt.ylabel("Sommes total des ventes ($)")
plt.title("Sommes des ventes par sexe")
plt.savefig("graphic.png", bbox_inches="tight")
plt.draw()
```

```python
plt.bar(achat_genre.index, achat_genre, color=["b", "r"], width=0.3)
plt.xlabel("Sexe")
plt.ylabel("Sommes total des ventes ($)")
plt.title("Sommes des ventes par sexe")
plt.savefig("graphic.png", bbox_inches="tight")
plt.draw()
```
